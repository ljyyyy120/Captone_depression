```python
import matplotlib.pyplot as plt

# Create figure with specified size
fig, ax = plt.subplots(figsize=(4, 1),dpi=900)

# Add scaling law formula as text
scaling_formula = r"$Y = Y_0 \cdot N^{\beta}$"
ax.text(0.5, 0.5, scaling_formula, fontsize=16, ha='center', va='center')

# Hide axes
ax.axis('off')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Set up figure
fig, ax = plt.subplots(figsize=(4, 1),dpi=900)
plt.rcParams.update({
    "text.usetex": False,
    "font.family": "Arial",
    "font.size": 14
})
# Define the equation in LaTeX format
equation = (
    r"$I = \left(\frac{n}{S_0}\right) \cdot "
    r"\frac{\sum_i \sum_j w_{ij}(e_{(i)} - \bar{e})(e_{(j)} - \bar{e})}"
    r"{\sum_i (e_{(i)} - \bar{e})^2}$" "\n\n"
    r"where:" "\n"
    r"  $n$ - number of spatial units" "\n"
    r"  $S_0$ - sum of all spatial weights" "\n"
    r"  $w_{ij}$ - spatial weight between units $i$ and $j$" "\n"
    r"  $e_{(i)}$ - residual at location $i$" "\n"
    r"  $\bar{e}$ - mean of residuals"
)


# Plot text in the center
ax.text(0.5, 0.5, equation, fontsize=16)

# Hide axes
ax.axis('off')

# Display
plt.tight_layout()
plt.show()
```
