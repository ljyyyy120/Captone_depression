```python
import pandas as pd
#import geopandas as gpd
import warnings
warnings.simplefilter(action='ignore')
```

```python
data_dir = "../00_data/01_raw/"

gdp = pd.read_csv(f"{data_dir}CAGDP1__ALL_AREAS_2001_2023.csv", encoding='latin-1')
us = pd.read_csv(f"{data_dir}CAGDP1_US_2001_2023.csv", encoding='latin-1')
```

```python
last_4_indices = gdp.index[-4:]  
gdp = gdp.drop(last_4_indices)
```

```python
gdp
```

```python
unique_descriptions = gdp['Description'].unique()
print(unique_descriptions)
```

```python
current_dollar_gdp = gdp[gdp['Description'] == 'Current-dollar GDP (thousands of current dollars) '] #2017
```

```python
us_fips_set = set(us["GeoFIPS"].unique())
current_dollar_gdp_filtered = current_dollar_gdp[~current_dollar_gdp["GeoFIPS"].isin(us_fips_set)]
```

```python
current_dollar_gdp_filtered["GeoFIPS"] = (
    current_dollar_gdp_filtered["GeoFIPS"].astype(str)
)
current_dollar_gdp_filtered["GeoFIPS"] = (
    current_dollar_gdp_filtered["GeoFIPS"]
    .str.replace('"', '', regex=False)   
    .str.strip()                  
)

current_dollar_gdp_filtered["GeoFIPS"] = (
    current_dollar_gdp_filtered["GeoFIPS"].str.zfill(5)
)

# 查看结果
current_dollar_gdp_filtered.head()
```

```python
year_columns = ["2011", "2012", "2013", "2014", "2015",
                "2016", "2017", "2018", "2019", "2020",
                "2021", "2022"]

df_long = current_dollar_gdp_filtered.melt(
    id_vars=["GeoFIPS"],       
    value_vars=year_columns,   
    var_name="year",            
    value_name="gdp"           
)

df_long = df_long[["year", "GeoFIPS", "gdp"]]
```

```python
df_long["gdp"] = (
    df_long["gdp"]
    .str.replace(",", "", regex=False)  # 如果有千位分隔逗号，先删除
    .str.strip()                        # 去除前后空格
)

# 然后将其转为数值
df_long["gdp"] = pd.to_numeric(df_long["gdp"], errors="coerce")
```

```python
df_long.to_csv('../00_data/12_intermediate/gdp.csv',index=False)
```
