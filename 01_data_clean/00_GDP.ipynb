```python
import pandas as pd
#import geopandas as gpd
import numpy as np
from statsmodels.tsa.stattools import adfuller
import statsmodels.formula.api as smf
from scipy.stats import pearsonr
import warnings
warnings.simplefilter(action='ignore')
```

```python
data_dir = "../00_data/01_raw/"
intermediate_dir = "../00_data/02_intermediate/"

df = pd.read_csv('https://data.nber.org/cbsa-csa-fips-county-crosswalk/cbsa2fipsxw.csv')
gdp = pd.read_csv(f"{data_dir}CAGDP1__ALL_AREAS_2001_2023.csv", encoding='latin-1')
us = pd.read_csv(f"{data_dir}CAGDP1_US_2001_2023.csv", encoding='latin-1')
```

```python
df
```

```python
last_4_indices = gdp.index[-4:]  
gdp = gdp.drop(last_4_indices)
```

```python
gdp
```

```python
unique_descriptions = gdp['Description'].unique()
print(unique_descriptions)
```

```python
current_dollar_gdp = gdp[gdp['Description'] == 'Current-dollar GDP (thousands of current dollars) '] #2017
```

```python
current_dollar_gdp
```

```python
us
```

```python
us_fips_set = set(us["GeoFIPS"].unique())
current_dollar_gdp_filtered = current_dollar_gdp[~current_dollar_gdp["GeoFIPS"].isin(us_fips_set)]
```

```python
df["fipsstatecode"] = df["fipsstatecode"].astype(str)
df["fipscountycode"] = df["fipscountycode"].astype(str)

df["fipsstatecode"] = df["fipsstatecode"].str.zfill(2) 
df["fipscountycode"] = df["fipscountycode"].str.zfill(3)
```

```python
df["countyfips"] = df["fipsstatecode"] + df["fipscountycode"]
```

```python
df
```

```python
current_dollar_gdp_filtered["GeoFIPS"] = (
    current_dollar_gdp_filtered["GeoFIPS"].astype(str)
)
current_dollar_gdp_filtered["GeoFIPS"] = (
    current_dollar_gdp_filtered["GeoFIPS"]
    .str.replace('"', '', regex=False)   
    .str.strip()                  
)

current_dollar_gdp_filtered["GeoFIPS"] = (
    current_dollar_gdp_filtered["GeoFIPS"].str.zfill(5)
)

# 查看结果
current_dollar_gdp_filtered.head()
```

```python
current_dollar_gdp_withcbsa = current_dollar_gdp_filtered.merge(
    df[["countyfips", "cbsacode"]],  
    how="left",                      
    left_on="GeoFIPS",             
    right_on="countyfips"          
)

current_dollar_gdp_withcbsa.drop(columns="countyfips", inplace=True)
current_dollar_gdp_withcbsa["cbsacode"] = current_dollar_gdp_withcbsa["cbsacode"].astype(str)
```

```python
current_dollar_gdp_withcbsa
```

```python
year_columns = ["2011", "2012", "2013", "2014", "2015",
                "2016", "2017", "2018", "2019", "2020",
                "2021", "2022"]

df_long = current_dollar_gdp_withcbsa.melt(
    id_vars=["cbsacode"],       
    value_vars=year_columns,   
    var_name="year",            
    value_name="gdp"           
)

df_long = df_long[["year", "cbsacode", "gdp"]]
```

```python
df_long = df_long.sort_values(by=["cbsacode", "year"])
df_long.head(20)
```

```python
df_long["gdp"] = (
    df_long["gdp"]
    .str.replace(",", "", regex=False)  # 如果有千位分隔逗号，先删除
    .str.strip()                        # 去除前后空格
)

# 然后将其转为数值
df_long["gdp"] = pd.to_numeric(df_long["gdp"], errors="coerce")
```

```python
df_cbsa_agg = (
    df_long
    .groupby(["cbsacode", "year"], as_index=False)  
    .agg({"gdp": "sum"})                       
)

print(df_cbsa_agg.head())
```

```python
df_cbsa_agg.to_csv(f'{intermediate_dir}gdp_cbsa.csv',index=False)
```
