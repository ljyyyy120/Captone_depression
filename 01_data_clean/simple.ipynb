# Merge again!

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
smart_path = '/Users/edison/Desktop/park/agg_smart_chr_new1.csv'
t19_path = '/Users/edison/Desktop/dp/twitter19_cbsa_depression_rates.csv'
t10_path = '/Users/edison/Desktop/dp/tweet2020_analysis.csv'
```

```python
smart_df = pd.read_csv(smart_path)
t19_df = pd.read_csv(t19_path)
t10_df = pd.read_csv(t10_path)
```

```python
smart_df['Data_value'] = smart_df['Data_value'].astype(float) / 100
smart_df = smart_df.drop(columns=['Data_value_unit'])
```

```python
smart_df['year'] = smart_df['year'].astype(int).astype(str)
print(smart_df['year'].unique())
```

```python
t10_df = t10_df.drop(columns=['cbsatitle'])
t10_df = t10_df.drop(columns=['metropolitanmicropolitanstatis'])
t10_df
```

```python
t10_df['cbsacode'] = t10_df['cbsacode'].astype(int)
print(t10_df['cbsacode'].unique())
```

```python
t19_df['cbsa'] = t19_df['cbsa'].astype(int)
print(t19_df['cbsa'].unique())
```

```python
t10_df['year'] = '2010'
```

```python
t10_df.rename(columns={'depression_rate': 'Data_value'}, inplace=True)
t19_df.rename(columns={'depression rate': 'Data_value'}, inplace=True)
t10_df.rename(columns={'cbsacode': 'cbsacode'}, inplace=True)
t19_df.rename(columns={'cbsa': 'cbsacode'}, inplace=True)
```

```python
smart_2011 = smart_df[smart_df['year'] == '2011'].copy()
smart_2011
```

```python
t10_merge_df = pd.merge(t10_df, smart_2011, on='cbsacode', suffixes=('', '_smart'), how='left')

t10_merge_df = t10_merge_df.drop(columns=['year_smart'])
t10_merge_df.rename(columns={'year': 'year'}, inplace=True)
```

```python
t10_merge_df
```

```python
t10_merge_df = t10_merge_df.drop(columns=['Data_value_smart'])
```

```python
smart_df = pd.concat([smart_df, t10_merge_df], axis=0)
```

```python
smart_df
```

```python
smart_2019 = smart_df[smart_df['year'] == '2019'].copy()
smart_2019
```

```python
t19_df['year'] = '2019'
```

```python
t19_merge_df = pd.merge(t19_df, smart_2019, on='cbsacode', suffixes=('', '_smart'), how='left')
```

```python
t19_merge_df = t19_merge_df.drop(columns=['Unnamed: 0'])
t19_merge_df = t19_merge_df.drop(columns=['year_smart'])
t19_merge_df = t19_merge_df.drop(columns=['population_smart'])
t19_merge_df = t19_merge_df.drop(columns=['Data_value_smart'])
t19_merge_df
```

```python
t19_merge_df['source'] = 'Twitter'
smart_df['source'] = 'Smart'

combined_df = pd.concat([smart_df, t19_merge_df], axis=0)
```
